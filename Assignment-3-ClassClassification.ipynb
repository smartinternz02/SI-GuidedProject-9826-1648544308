```python
# Imports
import pandas as pd
import numpy as np
import tensorflow as tf
import keras as ks

# SK Learn Specific Imports
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

# Keras Specific Imports
from keras.utils.np_utils import to_categorical
from keras.models import Sequential
from keras.layers import Dense
from keras import backend
```

```python
# Import DataSet
data = pd.read_csv("assign3_dataset.csv")
```

```python
# Original Data
data.head(5)
```

```python
# Filtered Data -> Not Nulls
nullval = pd.notnull(data['Position'])
data = data[nullval]
data.head()
```

```python
# Positions List
fwd = ["ST", "LW", "RW", "LF", "RF", "RS","LS", "CF"]
mid = ["CM","RCM","LCM", "CDM","RDM","LDM", "CAM", "LAM", "RAM", "RM", "LM"]
dfn = ["CB", "RCB", "LCB", "LWB", "RWB", "LB", "RB"]
```

```python
# Assign Values to Positions
data.loc[data["Position"] == "GK", "Position"] = 0
data.loc[data["Position"].isin(dfn), "Position"] = 1
data.loc[data["Position"].isin(mid), "Position"] = 2
data.loc[data["Position"].isin(fwd), "Position"] = 3
```

```python
# Count based on Values
data['Position'].value_counts()
```

```python
# Unique Values in Data
data['Position'].unique()
```

```python
#Filtered Data (34 of 89 Columns)
data = data[["Position", 'Finishing', 'HeadingAccuracy', 'ShortPassing', 'Volleys', 'Dribbling',
       'Curve', 'FKAccuracy', 'LongPassing', 'BallControl', 'Acceleration',
       'SprintSpeed', 'Agility', 'Reactions', 'Balance', 'ShotPower',
       'Jumping', 'Stamina', 'Strength', 'LongShots', 'Aggression',
       'Interceptions', 'Positioning', 'Vision', 'Penalties', 'Composure',
       'Marking', 'StandingTackle', 'SlidingTackle', 'GKDiving', 'GKHandling',
       'GKKicking', 'GKPositioning', 'GKReflexes']]
data.head()
```

```python
x = data.drop('Position',axis=1) # Drop Position Label from Filtered Data
```

```python
# Fit x -> StandardScaler to Panda's DataFrame
ss = StandardScaler()
x = pd.DataFrame(ss.fit_transform(x))
y = data['Position']
```

```python
x.head()
```

```python
y.head()
```

```python
y_cat = to_categorical(y)

y_cat[:10]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x.values, y_cat, test_size=0.2)
```

```python
y.shape
```

```python
x.shape
```

```python
backend.clear_session()
model = Sequential()
model.add(Dense(60, input_shape = (33,), activation = "relu"))
model.add(Dense(15, activation = "relu"))
model.add(Dense(4, activation = "softmax"))
model.summary()
```

```python
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
model.fit(x_train, y_train, verbose=1, epochs=25)
```

```python
y_pred = model.predict(x_test)
y_pred
```
